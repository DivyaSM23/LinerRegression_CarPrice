```python
num_df=df[['selling_price', 'year', 'km_driven','seats']]
```

```python
from pandas.api.types import CategoricalDtype 

def localonehot(df, feature):
    namelist  = df[feature].values.tolist()
    unamelist = list(set(namelist))
    pd.get_dummies(df[feature],prefix=feature,drop_first=True)
    df[feature]= df[feature].astype(CategoricalDtype(unamelist))
    name_df = pd.get_dummies(df[feature],prefix=feature)
    return name_df

name_df = localonehot(df, 'name')
year_df = localonehot(df, 'year')
fuel_df = localonehot(df, 'fuel')
st_df = localonehot(df, 'seller_type')
tras_df = localonehot(df, 'transmission')
owner_df = localonehot(df, 'owner')
engine_df = localonehot(df, 'engine')
max_power_df = localonehot(df, 'max_power')
torque_df = localonehot(df, 'torque')

```

```python
df1=pd.concat([num_df, name_df, year_df, fuel_df, st_df, tras_df, owner_df],axis=1)
df1
```

```python
target_df = df1[["selling_price"]]
print(target_df.shape)
target_df.head()
```

```python
input_df= df1.drop('selling_price', axis=1)
print(input_df.shape)
input_df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input_df, target_df, test_size=0.2, random_state=42, shuffle=True)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
y_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler()
X_train= sc.fit_transform(X_train)
X_test= sc.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
print(reg.coef_)
print(reg.intercept_)
pred = reg.predict( X_test)
```

```python
#from sklearn.metrics import confusion_matrix,accuracy_score
#reg.score(X_test,y_test)
```

```python
from sklearn.metrics import r2_score
score = r2_score( y_test, pred)
print("The accuracy of our model is {}%".format(round(score, 2) *100))
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dec_model= DecisionTreeRegressor()
dec_model.fit(X_train,y_train)
dec_model.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
random_model=  RandomForestRegressor()
random_model.fit(X_train,y_train)
random_model.score(X_test, y_test)
```
